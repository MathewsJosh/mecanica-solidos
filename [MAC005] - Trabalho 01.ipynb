image.png

### Uma transformação T opera em qualquer vetor $a$ para dar $Ta = \frac{a}{|a|}$, onde $|a|$ é a magnitude de $a$. Mostre que T não é uma transformação linear.

Como $Ta = \frac{a}{|a|}$ para todos os valores de $a$, então podemos fazer:

$T(a + b) =  \frac{(a + b)}{|a+b|}$

Com isso, escrevemos:

$Ta + Tb = \frac{a}{|a|} + \frac{b}{|b|}$

Nesse momento, já podemos notar que $T(a + b) \neq Ta + Tb$

Ou seja, T não é uma transformação linear.

### A) $Ta$

Usando as informações passadas no enunciado, podemos fazer

$Ta = T(2e1 + 3e2)$

$2Te1 + 3Te2 = 2(e1 + e2) + 3(e1 − e2)$

O que nos dá $5e1 -e2$. Ou seja, $Ta = 5e1 -e2$.

$Tb = T(3e1 + 2e2)$

$T(3e1 + 2e2) = 3Te1 + 2Te2$

Como $Te1 = e1 + e2$ e $Te2 = e1 − e2$, então fazemos:

$3Te1 + 2Te2 = 3(e1 + e2) + 2(e1 − e2)$

$3(e1 + e2) + 2(e1 − e2) = 3e1 + 3e2 + 2e1 - 2e2$

O que nos dá $5e1 + e2$. Ou seja, $Tb = 5e1 + e2$.

### B) $T(a+b)$

$T(a+b) = T(2e1 + 3e2 + 3e1 + 2e2)$

$T(a+b) = T(5e1 + 5e2)$

Dessa forma, temos $5(e1 + e2 + e1 − e2) = 5e1 + 5e1 + 5e2 - 5e2$

## Solução do problema 2.46 (inserir número e enunciado)


image.png

###a) Dado qualquer vetor $a$ e qualquer tensor $T$, mostrar que $a T^A a = 0$, onde $T^A$ e $T^S$ são simétricos e antisimétricos por parte de T.

R: Dado que $T^A$ é antissimétrico, então $(T^A)^T = -T^A$. Dessa forma, podemos calcular:

$aT^Aa = a(T^A)^Ta$ 

$aT^Aa = -aT^Aa$

$2aT^Aa = 0$

$aT^Aa = 0$

###b) Dado qualquer vetor $a$ e qualquer tensor $T$, mostrar que $a T a = a T^S a$, onde $T^A$ e $T^S$ são simétricos e antisimétricos por parte de T.

R: Dado que qualquer Tensor $T$ pode ser decomposto na soma de um tensor simétrico $T^S$ e um antissimétrico $T^A$. Temos: $T = T^S + T^A$

Sendo assim, podemos calcular:

$a T a = a(T^S + T^A)a$

$a T a = (a T^S + a T^A) a$

$a T a = a T^S a + a T^A a$

Substituindo o valor encontrado na alternativa a ($aT^Aa = 0$), temos: 

$a T a = a T^S a + 0$

$a T a = a T^S a$
