## 3.55

3.55.png

Em componentes cartesianas, a equação da conservação da massa, também conhecida como equação da continuidade, pode ser escrita da seguinte forma:

$p(\frac{\partial v_{1}}{\partial x_{1}} + \frac{\partial v_{2}}{\partial x_{2}} + \frac{\partial v_{3}}{\partial x_{3}}) + \frac{\partial p}{\partial t} + v_{1}\frac{\partial p}{\partial x_{1}} + v_{2}\frac{\partial p}{\partial x_[{2}} + v_{3}\frac{\partial p}{\partial x_{3}} = 0.$

Com os dados do enunciado, temos $v=α(x_{1}te_{1} + x_{2}te_{2})$ e podemos considerar:
$v_1 = αx_1t$ e $v_2 = αx_2t.$ Já $v_3$ é zero.

Agora podemos usar a equação da conservação da massa e os dados do enunciado para montar a integral e fazer os cálculos.

$\frac{dp}{dt} + p(αt + αt) = 0$

$→ \frac{dp}{dt} + p(2αt) = 0$

$→ \int_{p_0}^{p} \frac{dp}{p} = -2α \int_{0}^{t} t \,dt$

$→ ln \frac{p}{p_0} = - \alpha t^2$

A partir disso, podemos obter:

$→ \frac{p}{p_0} = e^{-αt^2}$

$→ p = p_0e^{-αt^2}$

Esse resultado determina como a densidade do fluido varia com o tempo.
